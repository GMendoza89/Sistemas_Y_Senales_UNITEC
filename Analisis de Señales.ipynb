## <center> Clasificación de Señales </center> ##

Las señales son manifestaciones de la amplia diversidad de fenómenos físicos que están presentes en nuestro entorno, 
en la naturaleza, en el planeta. 
Las señales también se producen de procesos y actividades realizados por el ser humano, 
en cualquier caso, contienen información y de ahí su importancia, 
cuanto más se conozca de una señal más se conocerá acerca del comportamiento y características del propio fenómeno físico o proceso en cuestión.

Las señales describen una gran diversidad de fenómenos físicos y su representación también es variada, ya sea en forma de datos, en forma gráfica o mediante una representación matemáticamente en función de una o más variables independientes. En cualquier caso, una clasificación general de señales es la siguiente:

* En Tiempo continuo y en Tiempo discreto.
* Par e Impar.
* Periódicas y No Periódicas.
* Determinísticas y Aleatorias.
* De Energía y de Potencia.

### Continuas y Discretas ###
Una señal es en tiempo continuo si está definida para una sucesión continua de valores de la variable independiente, en donde se asume que ésta es continua. 
La amplitud de las señales en tiempo continuo pueden adquirir valores continuos con el tiempo y tienen una notación de la forma $x(t)$
A continuación veremos ejemplos de señales continuas generadas a traves de Python. 
Primero necesitamos importar los módulos necesarios, 
en este caso van a ser dos, 
el módulo **numpy**, 
por medio del cual vamos a definir las matrices como arreglos, 
ademas de contar con funciones funciones trigonométricas y con funciones de multiplicación de matrices. 
El segundo módulo a utilizar es **pyplot** incluido en el paquete de **matplotlib**, 
el cual nos va ayudar a ver en una gráfica los punto necesarios.

```python
import numpy as np 
from matplotlib import pyplot as plt 
```

Ahora generamos los arreglos de las señales en el tiempo, con lo que graficaremos de -0.5 segundos a 0.5 segundo, generando datos cada 0.001 segundos, 
es decir vamos a generar un arreglo de 1000 datos que van a contener la información de 1 segundo.

```python
t = np.arange(-0.5, 0.5, 0.001)
wave1, wave2, wave3, wave4 = 5, 10, 50, 100
```

Vamos a generar las señales sinusoidales y una función compuesta pos la suma de las tres primeras. matemáticamente hablando tenemos las funciones como:

$f_{1} = \sin \left( 5 \pi t \right)$

$f_{2} = \sin \left( 10 \pi t \right)$

$f_{2} = \sin \left( 50 \pi t \right)$

$f =  \sin \left( 5 \pi t \right) + \sin \left( 10 \pi t \right) + \sin \left( 50 \pi t \right)$

```python
function1 = np.sin(wave1*np.pi*t)
function2 = np.sin(wave2*np.pi*t)
function3 = np.sin(wave3*np.pi*t)
functionCom1 = function1 + function2 + function3
```

Ahora vamos a utilizar el módulo de matplotlib para graficar las señales

```python
plt.subplot(121)
plt.ylabel("f(t)")
plt.xlabel("t")
plt.plot(t,function1,'r')
plt.plot(t,function2,'b')
plt.plot(t,function3,'g')
plt.grid()

plt.subplot(122)
plt.ylabel("f(t)")
plt.xlabel("t")
plt.plot(t,functionCom1)
plt.grid()
plt.savefig('Analog Waves.png', dpi = 300)
plt.show()
```

Las señales en tiempo discreto la variable independiente toma únicamente valores discretos, 
por lo que son continuas en amplitud y discretas en tiempo. 
Las señales en tiempo discreto se denotan como $ x[n] $ y de manera más específica como $x[nT_{s}]$, 
en donde $T_{s}$ corresponde a un intervalo o periodo de muestreo y $n$ representa la variable independiente de valor entero.
Las figuras siguientes generadas en Python, muestran diferentes señales discretas obtenidas al tomar muestras de una señal $x(t)$ seleccionando $Ts = .01s$, 
$Ts = .02s$, 
$Ts = .04s$ 
y $Ts = .08$s, 
respectivamente.

```python
n1 = np.arange(0, 0.6, 0.01)
n2 = np.arange(0, 0.6, 0.02)
n3 = np.arange(0, 0.6, 0.04)
n4 = np.arange(0, 0.6, 0.08)

function1 = np.sin(10*np.pi*n1)
function2 = np.sin(20*np.pi*n1)
functionCom1 = function1 + function2
plt.subplot(221)
plt.title("Ts = 0.01 s")
plt.ylabel("f[n]")
plt.xlabel("n")
plt.stem(n1,functionCom1)
plt.grid()

plt.subplot(222)
function3 = np.sin(10*np.pi*n2)
function4 = np.sin(20*np.pi*n2)
functionCom2 = function3 + function4 
plt.title("Ts = 0.02 s")
plt.ylabel("f[n]")
plt.xlabel("n")
plt.stem(n2,functionCom2)
plt.grid()

plt.subplot(223)
function5 = np.sin(10*np.pi*n3)
function6 = np.sin(20*np.pi*n3)
functionCom3 = function5+function6 
plt.title("Ts = 0.04 s")
plt.ylabel("f[n]")
plt.xlabel("n")
plt.stem(n3,functionCom3)
plt.grid()

plt.subplot(224)
plt.title("Ts = 0.08 s")
function7 = np.sin(10*np.pi*n4)
function8 = np.sin(20*np.pi*n4)
functionCom4 = function7 + function8 
plt.ylabel("f[n]")
plt.xlabel("n")
plt.stem(n4,functionCom4)
plt.grid()
plt.savefig('Discrete Waves.png')
plt.show()
```

Se observa que entre mayor es el periodo de muestreo Ts se va perdiendo la característica de la señal original.

## Señal Par e Impar ##

Una señal $x(t)$ o $x[n]$ es Par si, 
para toda $t$ o para toda $n$ es idéntica a su reflexión, 
inversión o transposición en el tiempo, 
es decir, 
es idéntica con respecto al eje de las ordenadas. 
En tiempo continuo es:

<center> $x(t) = x(-t)$ </center>

En tiempo discreto debe satisfacer la condición


<center> $x(n) = x(-n)$ </center>

```python
t = np.arange(-10.0, 10.0, 0.001)
mt = np.arange(10.0, -10.0, -0.001)
x1 = np.cos(np.pi*t/6)
x11 = np.cos(np.pi*mt/6)
x2 = np.sin(np.pi*t/6)
x21 = np.sin(np.pi*mt/6)
n = np.arange(-10.0, 10.0, 1)
mn = np.arange(10.0, -10.0, -1)
x3 = np.cos(np.pi*n/12)
x31 = np.cos(np.pi*mn/12)
x4 = np.sin(np.pi*n/12)
x41 = np.sin(np.pi*mn/12)

plt.figure(figsize=(16,9))

plt.subplot(221)
plt.title("Señal Par")
plt.plot(t,x1,'b',linewidth = 5)
plt.plot(t,x11,'r')
plt.grid()

plt.subplot(222)
plt.title("Señal Impar")
plt.plot(t,x2,'b')
plt.plot(t,x21,'r')
plt.grid()

plt.subplot(223)
plt.title("Señal Par")
plt.stem(n,x3,'b')
plt.stem(n,x31,'r')
plt.grid()

plt.subplot(224)
plt.title("Señal Impar")
plt.stem(n,x4,'b')
plt.stem(n,x41,'r')
plt.grid()

plt.tight_layout()
plt.show()
```

## Tranformación de señales ##

Un aspecto fundamental del análisis y procesamiento de señales y sistemas son las transformaciones, 
tanto en amplitud como en la variable independiente de tiempo, 
por lo que es conveniente presentar las transformaciones elementales que se pueden aplicar en una señal. 
De manera más específica, 
se puede modificar la amplitud, 
así como realizar tres operaciones sobre la variable independiente.


### Transformación de la señal en amplitud ###

Una señal $x(t)$ o $x[n]$ se puede modificar en amplitud con un valor A

<center> señal en tiempo continuo 

$x(t) = Ax(t)$ 

Señal en tiempo discreto

$x[t] = Ax[t]$</center>

<center>


Valor de $A$ | Trasformación en $x(t)$ 0 $x[n]$
------------ | --------------------------------
$\lvert A \rvert > 1$ | Se amplifica 
$ 0 < \lvert A \rvert < 1$ | Se atenúa
$ A < 0 $ | se invierte


</center>

Ejemplos de transformación en amplitud en TC con códigos en python,
en color azul se observa la señal fundamental sin variación la cual corresponde a:

<center>

$ x(t) = \cos \left(  t \right) $

</center>

En color rojo se observa la señal con la misma frecuencia, sin embargo, 
la amplitud esta multiplicada en un valor de $ A = 10 $, 
es decir,
la señal original amplificada.
La señal verde corresponde a la señal atenuada en un valor de 10 es decir $ A = \frac{1}{10} = 0.1$.
Por último la señal en color magenta corresponde a misma señal pero invertida es decir $ A = -1 $


```python
t = np.arange(-np.pi, np.pi,0.001)
A1 = 10.0
A2 = 0.1
A3 = -1.0
x = np.cos(5*t)
x1 = A1*x
x2 = A2*x
x3 = A3*x

plt.figure(figsize=(16,9))
plt.title("x(t)")
plt.plot(t,x,'b',linewidth = 2)
plt.plot(t,x1,'r',linewidth = 2)
plt.plot(t,x2,'g',linewidth = 2)
plt.plot(t,x3,'m',linewidth = 2)
plt.grid()
```

### Transformada de variable Independiente ###

Las transformaciones básicas de la variable independiente son 3

<center>

Transformación | Cambio en $ x(t)$ | Cambio en $ x[n]$ 
---------------|-------------------|------------------
Reflexión o inversión en el tiempo | $ (t) \rightarrow (-t)$  | $ [n] \rightarrow [-n]$ 
Desplazamiento en el tiempo  | $ (t) \rightarrow (t - t_{0})$  | $ [n] \rightarrow [n-n_{0}]$
Escalamiento en el tiempo  | $ (t) \rightarrow (\alpha t)$  | $ [n] \rightarrow [\alpha n]$  

</center>


Las tres transformaciones se pueden conjuntar originando una nueva señal modificada, para tiempo continuo:


<center>


$ x(t) \rightarrow x( \alpha t - t_{0})$


</center>


o bien para TD


<center>


$ x[n] \rightarrow x[\alpha n - n_{0} ]$


</center>




### Reflexión o inversión de tiempo ###

La reflexión de una señal $x(t)$ es $x(-t)$,
o bie la de  $ x[n]$ es $x[-n]$. Por ejemplo, si


<center>


$ x(t) = 10 \cos \left( 3t \right)$


</center>


entonces su reflexión es:

<center>


$ x(t) = 10 \cos \left( - 3t \right)$


</center>

En el  siguiente código analizaremos la señal $ x(t) = \sin(10 \pi t) + \sin(20 \pi t)$, 
generando la reflexión $ x(-t)$. 

```python
t = np.arange(-0.4, 0.3,0.001)
xt = np.sin(10*np.pi*t) + np.sin(20*np.pi*t)
x_t = np.sin(10*np.pi*(-t)) + np.sin(20*np.pi*(-t))

plt.figure(figsize=(16,9))

plt.subplot(311)
plt.title("x(t)")
plt.plot(t,xt,'b',linewidth = 2)
plt.grid()
plt.subplot(312)
plt.title("x(-t)")
plt.plot(t,x_t,'b',linewidth = 2)
plt.grid()
plt.subplot(313)
plt.title("x(t) y x(-t)")
plt.plot(t,xt,'b',linewidth = 2)
plt.plot(t,x_t,'r',linewidth = 2)
plt.grid()
```

Como se muestran en las imágenes la primera figura es la gráfica de la señal, 
la imagen siguiente es la reflexión de la señal, 
y en la figura 3 se conjuntan en la misma gráfica,
de esta forma es fácil observa que la señal $x(t)$ es una señal que no tiene paridad par. 

### Desplazamiento en el Tiempo ###

Una señal desplazada en tiempo corresponde a la misma señal pero trasladada un tiempo $t_{0}$ con respecto a la original. 
Se denota en tiempo continuo como $x(t − t_{0})$ y la correspondiente en tiempo discreto es $x[n − n_{0}]$. 
Si $t_{0}$ o $n_{0}$ son positivos entonces se dice que la señal esta atrasada. 
Por otro lado,
si son negativos entonces se tendrá una señal adelantada



```python
t = np.arange(-0.4, 0.3,0.001)
xt = np.sin(10*np.pi*t) + np.sin(20*np.pi*t)
t0 = 0.1
t1 = -0.05
xt_mt0 = np.sin(10*np.pi*(t - t0)) + np.sin(20*np.pi*(t - t0))
xt_Mt0 = np.sin(10*np.pi*(t - t1)) + np.sin(20*np.pi*(t - t1))



plt.figure(figsize=(16,9))
plt.title("x(t)")
plt.plot(t,xt,'b',linewidth = 2)
plt.plot(t,xt_mt0,'g',linewidth = 2)
plt.plot(t,xt_Mt0,'r',linewidth = 2)
plt.grid()
```

### Escalamiento de la Variable i dependiente t y n ###

El escalamiento en tiempo de una señal corresponde a comprimir o expandir la señal en el tiempo, 
es decir, 
se escala la variable independiente mediante cambios lineales en la misma.

Un ejemplo claro y sencillo de las transformaciones es en señales de voz. 
Para la señal $x(t)$ correspondiente a la palabra “camino”, 
$x(−t)$ corresponde a “onimac”. 
La señal comprimida $x(\alpha t)$ se escuchará más rápido mientras que la expandida $x \left( \frac{t}{\alpha} \right)$ se escucharía más lenta, 
$\alpha > 1$ representa el factor de escalamiento.
De manera general se puede expresar la señal escalada de $x(t)$ como $x(\alpha t)$, 
o bien, 
$x[n]$ como $x[\alpha n]$ donde $\alpha$ es una constante que representa el factor de escalamiento.

Si $\alpha > 1$, 
corresponde a una señal comprimida y por el contrario, si $ 0 < \alpha < 1$, 
representa una señal expandida.
Cuando $\alpha$ es negativa y diferente de 1 se obtiene una señal invertida y escalada en el tiempo.
